```python
from dotenv import load_dotenv
import os
from browserbase import Browserbase
from time import sleep

load_dotenv()

BROWSERBASE_API_KEY = os.environ["BROWSERBASE_API_KEY"]
BROWSERBASE_PROJECT_ID = os.environ["BROWSERBASE_PROJECT_ID"]

bb = Browserbase(api_key=BROWSERBASE_API_KEY, project_id=BROWSERBASE_PROJECT_ID)
```

```python
@bb.selenium
def get_title(driver):
    driver.get("https://www.browserbase.com")
    sleep(5)
    driver.get("https://kamath.io")
    return driver.title
    
get_title()
```

```python
session_id = bb.sessions['get_title']
bb.get_session_recording(session_id)
```
